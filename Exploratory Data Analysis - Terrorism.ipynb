## Step 1: Importing all required libraries

```python
import numpy as np, pandas as pd
import matplotlib as mpl, matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Step 2: Importing data

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv", encoding = "latin1")
df=pd.DataFrame(data)
print("Data has been successfully imported")
df.head()
```

## Step 3: Cleaning data

```python
df = df[["iyear", "imonth", "iday", "country_txt", "region_txt", "provstate", "city", "latitude", "longitude", "location",
         "summary", "attacktype1_txt", "targtype1_txt", "gname", "motive", "weaptype1_txt", "nkill", "nwound", "addnotes"]]
df.head()
```

```python
df.rename(columns={"iyear":"Year", "imonth": "Month", "iday": "Day", "country_txt": "Country", "region_txt": "Region",
                   "provstate": "Province/State", "city": "City", "latitude": "Latitude", "longitude": "Longitude",
                   "location": "Location", "summary": "Summary", "attacktype1_txt": "Attack Type", "targtype1_txt": "Target Type", 
                   "gname": "Group Name", "motive": "Motive", "weaptype1_txt": "Weapon Type", "nkill": "Killed",
                   "nwound": "Wounded", "addnotes": "Add Notes"}, inplace = True)
df.head()
```

## Step 4: Observing data

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["Killed"] = df["Killed"].fillna(0)
df["Wounded"] = df["Wounded"].fillna(0)
df["Casualty"] = df["Killed"] + df["Wounded"]
```

```python
df.describe()
```

### Observation of data

### Step 5: Visualizing data
### 1. Year wise attacks
#####   1. Number of attacks in each year

```python
Attacks = df["Year"].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index": "Year", "Year": "Attacks"}).set_index("Year")
Attacks.head()
```

```python
Attacks.plot(kind="bar", figsize = (15,6), fontsize = 10)
plt.title("Year wise Attacks", fontsize = 16)
plt.xlabel("Years", fontsize = 13)
plt.ylabel("Number of Attacks", fontsize = 13)
plt.show()
```

 1. Most number of attacks(16902) in 2014.
 2. Least number of attacks(471) in 1971.

##### 2. Total casualties ( killed + wounded ) in each year

```python
Casualties = df[["Year", "Casualty"]].groupby("Year").sum()
Casualties.head()
```

```python
Casualties.plot(kind = "bar", figsize = (15,6))
plt.title("Year wise Casulties", fontsize = 16)
plt.xlabel("Years", fontsize = 13)
plt.xticks(fontsize = 12)
plt.ylabel("Number of Casulties", fontsize = 13)
plt.show()
```

 1. Most number of attacks in 2014.
 2. Least number of attacks in 1970.

##### 3. Killed in each year

```python
Kill = df[["Year", "Killed"]].groupby("Year").sum()
Kill.head()
```

```python
Kill.plot(kind = "bar", figsize = (15,6))
plt.title("Year wise Killed", fontsize = 16)
plt.xlabel("Years", fontsize = 13)
plt.xticks(fontsize = 12)
plt.ylabel("Number of Killed", fontsize = 13)
plt.show()
```

### 2. Region wise attacks
##### 1. Distribution of terrorist attacks over regions from1970-2017

```python
Region_a = pd.crosstab(df.Year, df.Region)
Region_a.head()
```

```python
Region_a.plot(kind = "area", stacked = False, alpha = 0.5, figsize = (15,6))
plt.title("Region wise attacks", fontsize = 16)
plt.xlabel("Years", fontsize = 13)
plt.ylabel("Number of Attacks", fontsize = 13)
plt.show()
```

##### 2. Total terrorist attacks in each region

```python
Region_t = Region_a.transpose()
Region_t["Total"] = Region_t.sum(axis = 1)
ra = Region_t["Total"].sort_values(ascending  = False)
ra
```

```python
ra.plot(kind = "bar", figsize = (15,6))
plt.title("Total number of attacks in each region", fontsize = 16)
plt.xlabel("Regions", fontsize = 13)
plt.ylabel("Number of Attacks", fontsize = 13)
plt.show()
```

 1. Most number of attacks region: 50474 in Middle East & North Africa
